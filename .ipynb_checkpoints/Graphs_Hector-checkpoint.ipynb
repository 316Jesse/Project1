```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_data_df = pd.read_csv('chart samples csv.csv')


all_data_df
```

```python
all_data_df['House Price Index Central Florida'] = all_data_df['House Price Index Central Florida'] * 1000

```

```python
first_line = plt.plot(all_data_df["Year"], all_data_df["Mean Household Income"])
second_line = plt.plot(all_data_df["Year"], all_data_df["House Price Index Central Florida"])
plt.xlabel('Year', fontsize=11)
plt.ylabel('Sum of Mean Household Income / Sum of House Price CF', fontsize=9)
plt.ylim(0, 300000)
plt.xlim(1989, 2020)
plt.grid(True)
plt.show()
```

```python
plt.plot(all_data_df["Year"], all_data_df["Price Income Ratio"])


plt.xlabel('Year', fontsize=11)
plt.ylabel('Sum of Price Income Ratio', fontsize=11)
plt.ylim(1, 12)
plt.xlim(1989, 2020)
plt.title("Total")
plt.grid(True)

plt.show()
```
