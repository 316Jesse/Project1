```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
zori_df = pd.read_csv('zori_2023.csv')
zori_df
```

```python
new_zori_df = zori_df.loc[zori_df['StateName'] == "FL"]

new_zori_df = new_zori_df.drop(columns=['RegionID', 'SizeRank', 'RegionType'])

new_zori_df
```

```python
cf_zori_df = new_zori_df.loc[new_zori_df['RegionName'] == "Orlando, FL"]

cf_zori_df
```

```python
averages = cf_zori_df.mean()

average_dict = {"Average": averages}


central_FL_averages = pd.DataFrame(average_dict)
central_FL_averages = central_FL_averages.reset_index()



central_FL_averages
```

```python
central_FL_averages["index"]=pd.to_datetime(central_FL_averages["index"])



central_FL_averages["index"]= central_FL_averages["index"].dt.year


central_FL_averages
```

```python
year = central_FL_averages.groupby(['index'])
central_FL_averages = round(year.median(), 2)
central_FL_averages
```

```python
central_FL_averages['Average'] = central_FL_averages['Average'] * 12
central_FL_averages
```

```python
medianh_df = pd.read_csv('Zillow Median Sale Price.csv')
medianh_df
```

```python
new_medianh_df = medianh_df.loc[medianh_df['StateName'] == "FL"]

new_medianh_df = new_medianh_df.drop(columns=['RegionID', 'SizeRank', 'RegionType'])

new_medianh_df
```

```python
cf_medianh_df = new_medianh_df.loc[new_medianh_df['RegionName'] == "Orlando, FL"]

cf_medianh_df
```

```python
averages = cf_medianh_df.mean()

average_dict = {"Average": averages}

central_FLmh_averages = pd.DataFrame(average_dict)
central_FLmh_averages = central_FLmh_averages.reset_index()

central_FLmh_averages
```

```python
central_FLmh_averages["index"]=pd.to_datetime(central_FLmh_averages["index"])



central_FLmh_averages["index"]= central_FLmh_averages["index"].dt.year


central_FLmh_averages
```

```python
year = central_FLmh_averages.groupby(['index'])
central_FLmh_averages = round(year.mean(), 2)
central_FLmh_averages
```

```python
price_to_rent = central_FLmh_averages.div(central_FL_averages)

price_to_rent
```

```python
plt.figure(figsize=(10,6))
plt.plot(price_to_rent["Average"], label="Price to Rent Ratio")

plt.title('Price to Rent Ratio')
plt.ylim(5, 20)
plt.xlim(2018, 2022)
plt.xlabel('Year', fontsize=13)
plt.ylabel('Price to Rent Ratio', fontsize=13)
plt.legend(loc='best')
plt.grid(True)
```

```python
#Price-to-rent ratio of less than 15: It’s cheaper and more affordable to buy versus rent.
#Price-to-rent ratio of 16-20: Leans towards renting as a better option over buying.
#Price-to-rent ratio of over 21: By renting you are making a much better personal finance choice.
```
