```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
zori_df = pd.read_csv('zori_2023.csv')
zori_df
```

```python
new_zori_df = zori_df.loc[zori_df['StateName'] == "FL"]

new_zori_df = new_zori_df.drop(columns=['RegionID', 'SizeRank', 'RegionType'])

new_zori_df
```

```python
cf_zori_df = new_zori_df.loc[new_zori_df['RegionName'] == "Orlando, FL"]

cf_zori_df
```

```python
averages = cf_zori_df.mean()

average_dict = {"Average": averages}


central_FL_averages = pd.DataFrame(average_dict)
central_FL_averages = central_FL_averages.reset_index()



central_FL_averages
```

```python
central_FL_averages["index"]=pd.to_datetime(central_FL_averages["index"])



central_FL_averages["index"]= central_FL_averages["index"].dt.year


central_FL_averages
```

```python
year = central_FL_averages.groupby(['index'])
central_FL_averages = round(year.mean(), 2)
central_FL_averages
```
