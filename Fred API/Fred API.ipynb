```python
import fredapi as fa
import pandas as pd
from datetime import datetime as dt
from fred import api_key as key
fred = fa.Fred(key['api_key'])
#fred = fa.Fred('3258cc777cc97a53229e08f54d9846c1')
```

```python
#All-Transactions House Price Index for Florida
House_f = fred.get_series('ATNHPIUS36740Q')
House_f.name = 'House Price Index Central Florida' 
df0 = pd.DataFrame(House_f)
df0 = df0.reset_index() 
df0['index'] = df0['index'].dt.year
df0
```

```python
#rounding and taking the average for each year
year = df0.groupby(['index'])
df0 = round(year.mean(), 2)
df0.head()
```

```python
#Mean Adjusted Gross Income for Florida data pulled
Gross_income = fred.get_series('MEANAGIFL12A052NCEN')
Gross_income.name = 'Mean Household Income'
df00 = pd.DataFrame(Gross_income)
df00 = df00.reset_index() 
df00['index'] = df00['index'].dt.year
df00.head()
```

```python
#Mortgage intrest rates data pulled
Mortgage  = fred.get_series('MORTGAGE30US')
Mortgage.name = '30-Year Fixed Rate Mortgage Average'
df1 = pd.DataFrame(Mortgage)
df1 = df1.reset_index() 
df1['index'] = df1['index'].dt.year
df1.head()
```

```python
#rounding and taking the average for each year
year1 = df1.groupby(['index'])
df1 = round(year1.mean(),2)
df1.head()
```

```python
#Price House Index data pulled
House_index  = fred.get_series('FLSTHPI')
House_index.name = 'House Price Index for Florida'
df2 = pd.DataFrame(House_index)
df2 = df2.reset_index() 
df2['index'] = df2['index'].dt.year
df2
```

```python
#rounding and taking the average for each year
year2 = df2.groupby(['index'])
df2 = round(year2.mean(),2)
df2.head()
```

```python
#Unemployment in Florida data pulled
Unemployment   = fred.get_series('FLUR')
Unemployment.name = 'Unemployment Rate in Florida'
df3 = pd.DataFrame(Unemployment)
df3 = df3.reset_index() 
df3['index'] = df3['index'].dt.year
df3.head()
```

```python
#rounding and taking the average for each year
year3 = df3.groupby(['index'])
df3 = round(year3.mean(),2)
df3.head()
```

```python
#Private Housing Building Permits data pulled
permits  = fred.get_series('FLBP1FHSA')
permits.name = 'Private Housing Building Permits'
df4 = pd.DataFrame(permits)
df4 = df4.reset_index() 
df4['index'] = df4['index'].dt.year
df4.head()
```

```python
#rounding and taking the average for each year
year4 = df4.groupby(['index'])
df4 = round(year4.mean(),2)
df4.head()
```

```python
#Merging all the dataframes together
df5 = pd.merge(df00, df0, on='index')
df6 = pd.merge(df1, df5, on='index')
df7 = pd.merge(df2, df6, on='index')
df8 = pd.merge(df3, df7, on='index')
df9 = pd.merge(df4, df8, on='index')
#Putting the columns in a different order
df9 = df9[["index", "Mean Household Income", "House Price Index for Florida", 
           "House Price Index Central Florida","30-Year Fixed Rate Mortgage Average",
            "Unemployment Rate in Florida","Private Housing Building Permits"]]
#renaming the column index to year
df9 = df9.rename(columns={"index": "Year"})
#aligning all the columns and rows to make the data frame look nicer 
df9 = df9.style.set_table_styles(
[{'selector': 'th',
  'props': [('background', ''), 
            ('color', 'white'),
            ('font-family', 'verdana')]},
 
 {'selector': 'td',
  'props': [('font-family', 'verdana')]},

 {'selector': 'tr:nth-of-type(odd)',
  'props': [('background', '')]}, 
 
 {'selector': 'tr:nth-of-type(even)',
  'props': [('background', '')]},
 
]
).hide_index()
#printing the final dataframe
df9
```
