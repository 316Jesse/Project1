```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
#read full csv file
df = pd.read_csv('chart samples csv.csv')

#drop empty rows
df = df.dropna()

#change Year to integer
df['Year'] = df['Year'].astype(int)

#set year as index
df.set_index('Year', inplace=True)

#show dataframe
df.head()
```

```python
#remove % signs
df['Mean House Income Growth'] = df['Mean House Income Growth'].str.replace('%', '')
df['House Price Index FL Growth'] = df['House Price Index FL Growth'].str.replace('%', '')
df['House Price Index CF Growth'] = df['House Price Index CF Growth'].str.replace('%', '')
df['Private Housing Permits Growth'] = df['Private Housing Permits Growth'].str.replace('%', '')
df['Unemployment Rate in Florida.1'] = df['Unemployment Rate in Florida.1'].str.replace('%', '')

#change type to float
df['Mean House Income Growth'] = df['Mean House Income Growth'].astype(float)
df['House Price Index FL Growth'] = df['House Price Index FL Growth'].astype(float)
df['House Price Index CF Growth'] = df['House Price Index CF Growth'].astype(float)
df['Private Housing Permits Growth'] = df['Private Housing Permits Growth'].astype(float)
df['Unemployment Rate in Florida.1'] = df['Unemployment Rate in Florida.1'].astype(float)


df.head()
```

```python
#set figure size and plots lines
plt.figure(figsize=(10,6))
plt.plot(df['Mean House Income Growth'], label="Mean FL Household Income Growth")
plt.plot(df['House Price Index FL Growth'], label="House Price Index FL Growth")
plt.plot(df['House Price Index CF Growth'], label="House Price Index CF Growth")
plt.plot(df['Private Housing Permits Growth'], label="Private Housing Permits Growth")

#chart edits
plt.title('Growth % Rates')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100.0))
plt.xlabel('Year')
plt.ylabel('Percent Change')
plt.legend(loc='best')
plt.grid()
plt.savefig("Project 1 Charts/Growth % Rates (1989-2023).png")
plt.show()
```

```python
#set figure size and plots lines
plt.figure(figsize=(10,6))
plt.plot(df['Mean House Income Growth'], label="Mean FL Household Income Growth")
plt.plot(df['Unemployment Rate in Florida.1'], label="FL Unemployment Rate")

#chart edits
plt.title('Mean Household Income Growth vs. FL Unemployment (1989-2023)')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100))
plt.xlabel('Year')
plt.ylabel('Percent Change')
plt.legend(loc='best')
plt.grid()
plt.savefig("Project 1 Charts/Mean Household Income Growth vs. FL Unemployment.png")
plt.show()
```

```python
#set figure size and plots lines
plt.figure(figsize=(10,6))
plt.plot(df['House Price Index for Florida'], label="House Price Index for Florida")
plt.plot(df['House Price Index Central Florida'], label="House Price Index Central Florida")

#chart edits
plt.title('House Price Index for FL and Central FL (1989-2023)')
plt.xlabel('Year')
plt.ylabel('Index')
plt.legend(loc='best')
plt.grid()
plt.savefig("Project 1 Charts/House Price Index for FL and Central FL (1989-2023).png")
plt.show()
```
