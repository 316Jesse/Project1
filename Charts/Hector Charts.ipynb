```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
#Reading File
chart_df = pd.read_csv('chart samples csv.csv')
chart_df.set_index('Year', inplace=True)


chart_df
```

```python
#First Chart
plt.figure(figsize=(20,13))
plt.plot(chart_df["Price Income Ratio"], label="Price Income Ratio", marker='o')

for i, value in enumerate(chart_df["Price Income Ratio"].values):
    plt.annotate(str(value), xy=(chart_df.index[i], value), xytext=(-10, 10), textcoords='offset points', ha='right', va='bottom')

plt.title('Price Income Ratio (1989 - 2022)')
plt.ylim(1, 12)
plt.xlim(1989, 2022)
plt.xlabel('Year', fontsize=13)
plt.ylabel('Price Income Ratio', fontsize=13)
plt.legend(loc='best')
plt.grid(True)

plt.savefig("Project 1 Charts/Price Income Ratio Chart")
plt.show()
```

```python
chart_df['House Price Index Central Florida'] = chart_df['House Price Index Central Florida'] * 1000
```

```python
#Second Chart
plt.figure(figsize=(10,6))
plt.plot(chart_df["Mean Household Income"], label="Mean Household Income")
plt.plot(chart_df["House Price Index Central Florida"], label="House Prices Central Florida")

plt.title('Mean Household Income for Florida / House Prices Central Florida Comparison (1989 - 2022)')
plt.xlabel('Year', fontsize=11)
plt.ylabel('Mean Household Income for Florida / House Prices Central Florida', fontsize=9)
plt.ylim(0, 300000)
plt.xlim(1989, 2022)
plt.grid(True)
plt.legend(loc='best')
plt.savefig("Project 1 Charts/Mean Household Income_House Prices Central Florida Comparison")
plt.show()
```
